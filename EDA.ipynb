```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("s3://udemy-ds-lab/house_rentals.csv")

df.head()
```

```python
df.describe(include='all')
```

```python
plt.hist(df['ADS_PRICE'], bins=30, edgecolor='black', alpha=0.7)
plt.title('Distribution of ADS_PRICE')
plt.xlabel('ADS_PRICE')
plt.ylabel('Frequency')
plt.show()
```

```python
import seaborn as sns 

corr = df.corr()
sns.heatmap(corr, annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap')
plt.show()
```

```python
plt.figure(figsize=(12, 6))
sns.boxplot(x='ADS_PRICE', y='ADS_CATEGORY_NAME', data=df)
plt.title('Boxplot of ADS_PRICE vs. ADS_CATEGORY_NAME')
plt.xlabel('ADS_PRICE')
plt.ylabel('ADS_CATEGORY_NAME')
plt.xticks(rotation=45)
plt.show()
```

```python
sns.scatterplot(x='ADS_ATTR_SQUARE', y='ADS_PRICE', data=df)
plt.title('Scatterplot of ADS_PRICE vs. ADS_ATTR_SQUARE')
plt.xlabel('ADS_ATTR_SQUARE')
plt.ylabel('ADS_PRICE')
plt.show()
```

```python
sns.countplot(x='ADS_ATTR_ROOMS', data=df)
plt.title('Countplot of ADS_ATTR_ROOMS')
plt.xlabel('ADS_ATTR_ROOMS')
plt.ylabel('Count')
plt.show()
```
